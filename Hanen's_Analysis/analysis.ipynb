```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Name of the CSV file
file1 = 'Ressources/Coffee_Comsumption.csv'
file2 = "Ressources/State_Depression_Year.csv"
comsumption_df = pd.read_csv(file1)
depression_df = pd.read_csv(file2)
comsumption_df.head()
```

```python
depression_df.head()
```

```python
#Merging Consumption dataframe and corosponding Depression percentage dataframe
merge_df = pd.merge(comsumption_df,depression_df, on=["State","Year"], how="left")
merge_df
```

```python
#Grouping by the year for each state and average of Depression Percentage
group_df = merge_df.groupby("Year").mean(numeric_only=True)
plt.plot(group_df.index, group_df["Depression_Percentage"])
plt.xlabel("Year")
plt.ylabel("Depression Percentage (%)")
plt.title("Depression Percentage Per Year")
plt.show()
```

```python
#Grouping by the year for each state and average of ALC RATE
group_df = merge_df.groupby("Year").mean(numeric_only=True)
plt.plot(group_df.index, group_df["ALC Rate (%)"])
plt.xlabel("Year")
plt.ylabel("ALC Rate")
plt.title("ALC Rate Per Year")
plt.show()
```
