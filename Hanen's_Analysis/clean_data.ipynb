```python
df.reset_index(drop=True, inplace=True)
```

```python
df = df.rename(columns={"Years":"State"})
```

```python
col_change= ["2011", "2012","2013","2014","2015","2016","2017"]
df[col_change] = df[col_change].astype("float64")
df.dtypes
```

```python
df.loc["2011":"2020"] = df.loc["2011":"2020"]*100
df.head()
```

```python
df = df.set_index("State")
df = df.iloc[1:]
df
```

```python
df_long = df.reset_index().melt(id_vars=['State'], var_name='Year', value_name='Depression_Percentage')
```

```python
df_long
```

```python
df_long = df_long.set_index('State')
df_long
```

```python
df_long["Depression_Percentage"] = df_long["Depression_Percentage"]*100
df_long
```

```python
df_long = df_long.sort_index()
df_long 
```

```python
df_long.index = df_long.index.str.replace(r' \(\%\)', '', regex=True).str.replace(' ','')
df_long
```

```python
df_long.to_csv("Ressources/State_Depression_Year.csv")
```
