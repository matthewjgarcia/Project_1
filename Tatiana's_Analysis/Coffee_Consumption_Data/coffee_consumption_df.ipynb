```python
import pandas as pd
from pathlib import Path
```

```python
#Create a file path for the coffee consumption csv file

file = Path('Resources/Coffee_domestic_consumption.csv')
```

```python
#Read the csv in pandas in order to create a pandas df

coffee_consumption_df = pd.read_csv(file)

coffee_consumption_df.head()
```

```python
#Delete unecessary columns 
#We only need years 2010-2020, so remove the rest
#In thousand 60kg bags


coffee_consumption_df = coffee_consumption_df.loc[:,["Country", "Coffee type", "2010/11", "2011/12", "2012/13", "2013/14", "2014/15", "2015/16", "2016/17", "2017/18", "2018/19", "2019/20"]]


coffee_consumption_df
```

```python
coffee_consumption_df.to_csv("Resources/coffee_global_consumption.csv", sep=',', index=False, encoding='utf-8')
```
